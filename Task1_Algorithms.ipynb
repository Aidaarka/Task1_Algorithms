4.1 Naive method

```python
# using naive method
def naive(poly,x):
    naive_poly_result = 0
    for i in range(len(poly)):
        naive_poly_result += poly[i]*x**i 
    return naive_poly_result

#List of average time
total_time = []

for n in range(1, 1700):
    
    timer=[]
    
    for i in range(5):
        
#       Initialize Vector
        vector = np.array([random.randint(0, 100) for m in range(n)])
        x = 1.5
        
#       Start of time
        startTime = time.perf_counter()
        
#       Naive method
        naive_method = naive(vector, x)

                    
#       End of time
        endTime = time.perf_counter()
#       Calculation time
        timer.append(endTime - startTime)

    
#   Average time
    avg_time = mean(timer)
    total_time.append(avg_time)

fig = plt.figure(figsize = (12, 10))
plt.plot([n for n in range(len(total_time))], total_time, linewidth = 2.0)

plt.title('Calculation time of the Naive method')
plt.xlabel('Size of n')
plt.ylabel('Time; sec')

plt.grid()
plt.show() 
```

```python
# using Horner's method
def horner(poly, x):
 
    horner_poly_result = poly[0] 
  
    for i in range(len(poly)):
        horner_poly_result = horner_poly_result*x + poly[i]
    return horner_poly_result

#List of average time
total_time = []

for n in range(1, 2001):
    

    timer=[]
    
    for i in range(5):
        
#       Initialize Vector
        vector = np.array([random.randint(0, 100) for m in range(n)])
        x = 1.5
        
#       Start of time
        startTime = time.perf_counter()
        
#       Hornerâ€™s method
        horner_method = horner(vector, x)

                    
#       End of time
        endTime = time.perf_counter()
#       Calculation time
        timer.append(endTime - startTime)
    
#   Average time
    avg_time = mean(timer)
    total_time.append(avg_time)

fig = plt.figure(figsize = (12, 10))
plt.plot([n for n in range(len(total_time))], total_time, linewidth = 2.0)

plt.title('Calculation time of the Horners method')
plt.xlabel('Size of n')
plt.ylabel('Time; sec')

plt.grid()
plt.show() 
```

5. Bubble Sort

```python
#List of average time
total_time = []

for n in range(1, 500):
    
    timer=[]
    
    for i in range(5):
        
#       Initialize Vector
        vector = np.array([random.randint(0, 100) for m in range(n)])
        
#       Start of time
        startTime = time.perf_counter()
        
#       Bubble Sort
        for i in range(len(vector) - 1):
            for j in range(len(vector) - i - 1):
                if vector[j] > vector[j+1]:
                    vector[j], vector[j+1] = vector[j+1], vector[j]
                    
#       End of time
        endTime = time.perf_counter()
#       Calculation time
        timer.append(endTime - startTime)

    
#   Average time
    avg_time = mean(timer)
    total_time.append(avg_time)

fig = plt.figure(figsize = (12, 10))
plt.plot([n for n in range(len(total_time))], total_time, linewidth = 2.0)

plt.title('Calculation time of the Bubble Sort algorithm')
plt.xlabel('Size of n')
plt.ylabel('Time; sec')

plt.grid()
plt.show() 
```
